```python
import bs4
import requests
```

```python
r = requests.get('https://www.wunderground.com/forecast/VTCR%E2%80%9D')
```

```python
r.text
```

```python
html_page = bs4.BeautifulSoup(r.text, 'html.parser')
print(html_page)
```

```python
titles = html_page.findAll('span', {'class':'wu-value wu-value-to'} )
print(titles)
```

```python
title_list = []
for t in titles:
    obj = t.string 
    title_list.append(obj)
```

```python
print(title_list)
```
